```python
#Importing Python Packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score

#Importing All VI CSV's In Chronological Order:

#2020
date1 = pd.read_csv("C:/NSF-REU/Height_Combined_Files/2020/June_18_2020_Combined.csv")
date2 = pd.read_csv("C:/NSF-REU/Height_Combined_Files/2020/June_23_2020_Combined.csv")
date3 = pd.read_csv("C:/NSF-REU/Height_Combined_Files/2020/June_30_2020_Combined.csv")
date4 = pd.read_csv("C:/NSF-REU/Height_Combined_Files/2020/July_6_2020_Combined.csv")

#2021
date5 = pd.read_csv("C:/NSF-REU/Height_Combined_Files/2021/June_16_2021_Combined.csv")
date6 = pd.read_csv("C:/NSF-REU/Height_Combined_Files/2021/June_22_2021_Combined.csv")
date7 = pd.read_csv("C:/NSF-REU/Height_Combined_Files/2021/June_30_2021_Combined.csv")
date8 = pd.read_csv("C:/NSF-REU/Height_Combined_Files/2021/July_12_2021_Combined.csv")

#2022
date9 = pd.read_csv("C:/NSF-REU/Height_Combined_Files/2022/June_15_2022_Combined.csv")
date10 = pd.read_csv("C:/NSF-REU/Height_Combined_Files/2022/June_23_2022_Combined.csv")
date11 = pd.read_csv("C:/NSF-REU/Height_Combined_Files/2022/June_28_2022_Combined.csv")
date12 = pd.read_csv("C:/NSF-REU/Height_Combined_Files/2022/July_13_2022_Combined.csv")

date9 = date9.drop(columns=['id_old'])
date10 = date10.drop(columns=['id_old'])
date11 = date11.drop(columns=['id_old'])
date12 = date12.drop(columns=['id_old'])

dates = [date1, date2, date3, date4, date5, date6, date7, date8, date9, date10, date11, date12]
columns = ['Red','Green','Blue','BI','GLI','NGRDI','VARI','BGI','ExG','ExR','ExB','ExGR','MGRVI','RGBVI','GRRI','VEG','Elevation(','Moisture(%','Height']
df = pd.DataFrame(columns)

def feature_importance():
    for date in dates:
        key_features = date.drop(columns=['Plot', 'Year', "Date", 'Row', 'Range', 'left', 'top', 'right', 'bottom', 'Yld Mass(D', 'Yld Vol(Dr', 'Crop Flw(M', 'Crop Flw(V', 'Mean.Yld.bu.ac', 'X', 'Y'])
    for i in range(0, df.last_valid_index()+1):
        x = key_features.drop(columns=[columns[i]])
        y = date['Mean.Yld.bu.ac']
        x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.70)
        model = RandomForestRegressor(n_estimators=100) 
        model.fit(x_train, y_train)
        y_predict = model.predict(x_test)
        print("R-Squared:", r2_score(y_test, y_predict))
        importances = model.feature_importances_
        field_importance = pd.Series(importances, index=x.columns)
        field_importance.plot(kind='bar')
        plt.bar(range(len(model.feature_importances_)), model.feature_importances_)
        plt.yticks(np.arange(0, importances.max(), 0.1)) 
        plt.rc('xtick', labelsize=20) 
        plt.rc('ytick', labelsize=25)
        plt.show()
        print(importances, "\n")

feature_importance()
```
